## Ejercicio 1.1

* Genera una función que nos diga si un número es múltiplo de otro. 
* Ponla a prueba pidiendo por pantalla al usuario que introduzca los dos números al llamar al método.
* Prepara la petición por pantalla para que si el usuario no introduce números enteros y distintos del 0 le pida que vuelva a introducirlo, poniéndolo nuevamente a prueba.

```python
# Función que nos indique si un número es múltiplo de otro: el número se puede dividir sin dejar resto.

def es_multiplo(numero, multiplo):
    # Si el residuo es 0, es múltiplo
    if numero % multiplo == 0:  # operador módulo (%):nos devuelve el resto de la división de dos números, si es cero son divisibles entre sí.
        return True
    else:
        return False
```

```python
# Podemos simplifcar la función de esta forma:

def multiple(x, y):  
  return x % y == 0
```

```python
# Llamamos a la función:

es_multiplo(5,5) # es múltiplo
```

```python
es_multiplo(10,11) # no es múltiplo
```

```python
# Llamamos a la funciòn simplificada:

multiple(1,2)
```

```python
multiple(6,6)
```

```python
# Prueba de la función ingresando un número:

input1 = int(input("Ingrese el primer número: ")) # introducimos el 5, por ejemplo
input2 = int(input("Ingrese el segundo número: ")) # introducimos el 8

if(multiple(input1, input2)): 
  print("El número {0} es múltiplo de {1}".format(input1, input2)) 
else: 
  print("El número {0} no es múltiplo de {1}".format(input1, input2))
```

## Ejercicio 1.3

Vamos a trabajar ahora creando un montón de funciones para poder trabajar con fracciones de la forma a/b. Esto yo lo he usado para crear clases, pero me parece más útil que lo usemos como varios ejercicios de funciones seguidos, que tampoco tenéis tanto tiempo. Para representar una fracción vamos a utilizar dos enteros: numerador (la 'a') y un denominador (la 'b').  

Así pues, nuestro trabajo se basa en conseguir generar las siguientes funciones, que no son pocas:  

* leer_fraccion: Queremos que el usuario introduzca 'a' y 'b' por pantalla. Además, queremos que simplifiques la fracción. Para ello tendrás que crear otra función llamada simplificar_funcion, que haréis en el siguiente punto. 
* simplificar_funcion: Esto se hace dividiendo 'a' y 'b' entre el máximo común divisor (MCD) de ellos dos. Os daré esa función para ir más rápido. La idea es que esta función nos devuelva los dos valores ya arreglados.
* escribir_fraccion: Ahora queremos simplemente que nos imprima la fracción de la forma 'a/b', pero como sabéis, si el denominador es 1, entonces con que se imprima 'a' es suficiente. 
* sumar_fracciones: Dadas dos fracciones 'a/b' y 'c/d', queremos sumarlas. Para ello hay que recurrir a las fórmulas que vimos hace años en secundaria, que tenéis por Google o que os dejo aquí delante: numerador = a·d+b·c, denominador = b·d. Por supuesto debemos simplificar nuevamente el resultado antes de devolverlo.
* restar_fracciones: Literalmente lo que acabáis de hacer pero cambiando el signo '+' por el signo '-'. 
* multiplicar_fracciones: Aquí nos toca multiplicar numeradores por un lado, denominadores por otro y luego simplificar el resultado.
* dividir_fracciones: Dividir entre una fracción es multiplicar por la inversa. Así, el numerador tendrá que ser a·d y el denominador será b·c. Por supuesto toca simplificar el resultado.   

Finalmente comprobad que funcionan estas funciones. 

```python
# Funciones:

# leer_fraccion: Queremos que el usuario introduzca 'a' y 'b' por pantalla.

def leer_fraccion():
  a = int(input("Introduzca el numerador de la fracción: "))
  b = int(input("Introduzca el denominador de la fracción: "))
  return (a, b)
```

```python
# simplificar_funcion: Esto se hace dividiendo 'a' y 'b' entre el máximo común divisor (MCD) de ellos dos:

# Calculamos el MCD con esta funcion:


def mcd(a, b):   
  # Esta función calcula el Máximo Común Divisor (MCD) de dos enteros. 

  while b != 0: # bucle while que se ejecuta mientras b sea distinto de 0. 
    a, b = b, a % b # asigna a los dos enteros el valor de b y el resto de dividir a entre b
  return a # devuelve el valor de a como el MCD.
```

```python
# hacemos la función que lo ejecute:

def dividir_por_mcd(a, b):
  divisor = mcd(a, b)
  return (a / divisor, b / divisor)
```

```python
# Vamos a ver si funciona:

mcd(2,4) # calculamos el MCD
```

```python
dividir_por_mcd(2,4)
```

```python
# escribir_fraccion: Ahora queremos simplemente que nos imprima la fracción de la forma 'a/b', pero como sabéis, 
# si el denominador es 1, entonces con que se imprima 'a' es suficiente.

def escribir_fraccion(a, b):
  if b == 1:
    print(a)
  else:
    print(f"{a}/{b}")
```

```python
escribir_fraccion(1,2) # probamos que nos devuelve la función
```

```python
escribir_fraccion(1,1)
```

```python
# sumar_fracciones: Dadas dos fracciones 'a/b' y 'c/d', queremos sumarlas.

def sumar_fracciones(a, b, c, d):
  numerador = a * d + c * b
  denominador = b * d
  return (numerador, denominador)
```

```python
# probamos la función: 

sumar_fracciones(1,2,3,4)
```

```python
# restar_fracciones: 

def restar_fracciones(a, b, c, d):
  numerador = a * d - c * b
  denominador = b * d
  return (numerador, denominador)
```

```python
# probamos la función:

restar_fracciones(9,8,4,5)
```

```python
# multiplicar_fracciones:

def multiplicar_fracciones(a, b, c, d):
  numerador = a * c
  denominador = b * d
  return (numerador, denominador)
```

```python
multiplicar_fracciones(3,5,6,7) # probamos la función
```

```python
# dividir_fracciones: 

def dividir_fracciones(a, b, c, d):
  numerador = a * d  #multiplicamos a la inversa
  denominador = b * c
  return (numerador, denominador)
```

```python
# llamamos a la función:

dividir_fracciones(6,9,3,5)
```

```python
# matriz de 4x5:

arr = np.array(arr).reshape(4, 5)
print(arr)
```

```python
# Convertimos el tipo de dato  a entero. 
np.random.seed(15)
arr = np.random.normal(20, 9, 20)

arr = [int(round(x, 2)) for x in arr] # convertimos el num a entero
arr = np.array(arr).reshape(4, 5)

print(arr)
```

```python
# calculamos la media por columnas

means = np.mean(arr, axis=0) 

print(means)
```

## Ejercicio 4.2

* Dibuja en 3D los valores que están en el intervalo $[-2,2]$ en ambas variables para la función $Z=X^2-Y^2$. Como curiosidad, esta forma se llama "monkey saddle". 
* Añade cierto error vertical a cada punto.

```python
# para dibujar en 3D:
import pandas as pd
from mpl_toolkits.mplot3d import axes3d # para gráfIcos 3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Dibuja en 3D los valores que están en el intervalo [-2,2] en ambas variables para la función Z=X^2-Y^2.

fig = plt.figure(figsize=(14,6))

x = np.linspace(-2, 2)
y = np.linspace(-2, 2)

X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2

ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)


ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

```python
# Añade cierto error vertical a cada punto:

fig = plt.figure(figsize=(14,6))

x = np.linspace(-2, 2)
y = np.linspace(-2, 2)

X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2
Z += np.random.normal(0, 0.1, Z.shape) # Añadimos cierto error vertical a cada punto
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)


ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

```python
import pandas as pd
from sklearn import datasets

# cargamos el dataset de iris:

data=datasets.load_iris()
```

```python
 # Usamos el Test de Shapiro-Wilk para comprobar si las variables que tenemos siguen una distribución normal:

from scipy.stats import shapiro

shapiro_test = shapiro(df['sepal length (cm)'])

# Mostramos resultado

print(f"El p-valor obtenido en el test de Shapiro-Wilk es de {shapiro_test[1]}")
```

```python
shapiro_test = shapiro(df['sepal width (cm)'])

# Mostramos resultado

print(f"El p-valor obtenido en el test de Shapiro-Wilk es de {shapiro_test[1]}")
```

```python
shapiro_test = shapiro(df['petal length (cm)'])

# Mostramos resultado

print(f"El p-valor obtenido en el test de Shapiro-Wilk es de {shapiro_test[1]}")
```

```python

shapiro_test = shapiro(df['petal width (cm)'])

# Mostramos resultado

print(f"El p-valor obtenido en el test de Shapiro-Wilk es de {shapiro_test[1]}")
```

Si el p-valor es mayor que 0.05, podemos concluir que los datos siguen una distribución normal, en este caso tres de ellas lo cumplen. Si el p-valor es menor que 0.05, entonces concluimos que los datos no siguen una distribución normal, es el caso de 'sepal length (cm)'.

```python
# cargamos el dataset

df = sns.load_dataset('seaice')
```

```python
# ¿Tendría sentido realizar un estudio detallado sobre la serie temporal? Para ello, necesitaríamos que tuviese cierta regresión hacia su media.

# Para esto, se puede utilizar la prueba Dickey-Fuller para probar si hay una tendencia estacionaria en la serie temporal. 
# Si el p-value de la prueba es menor que 0.05 entonces se puede decir que hay una tendencia estacionaria en la serie temporal.

import statsmodels.tsa.stattools as ts

# realizar prueba Dickey-Fuller
tstat, pval, _, _, _, _ = ts.adfuller(df['Extent'])

# imprimir resultados
print('t-statistic:', tstat)
print('p-value:', pval)

# interpretar resultados
if pval < 0.05:
    print('Hay una tendencia estacionaria en la serie temporal.')
else:
    print('No hay una tendencia estacionaria.')
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
# Escalado de datos
from sklearn.preprocessing import StandardScaler
# Modelos
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
tips = sns.load_dataset('tips') # cargamos el dataset 
```

```python
tips.head()
```

```python
# Para realizar una regresión, primero necesitamos definir las variables independientes (X) y dependientes (y). 
# En este caso, las variables independientes son 'total_bill' y 'size' y la variable dependiente es 'tip'.

# Asignamos las columnas pertinentes a X e y:

X = tips[['total_bill', 'size']]
y = tips['tip']
```

```python
# Dividimos los datos en entrenamiento y test:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# Instaciamos el modelo:

model = LinearRegression()
```

```python
# Entrenamos el modelo:

model.fit(X_train, y_train)
```

```python
# Predice con el modelo
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculamos el error cometido por las predicciones:

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse) #calculamos la raíz cuadrada del error cuadrático medio (MSE)
mae = mean_absolute_error(y_test, y_pred)

# Muestra los errores
print('MSE: ', mse)
print('RMSE: ', rmse)
print('MAE: ', mae)
```
